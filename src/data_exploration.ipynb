```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm

path = '/datax/scratch/zachy/ml-blc1/data'

def plot_img(img):
    plt.imshow(np.load(img), aspect='auto', interpolation='none'); plt.colorbar()

def save_img(path):
    for root, dirs, files in os.walk(path):
        for filename in files:
            if filename != 'labels.csv':
                img = os.path.join(root, filename)
                plt.imsave(f'{root}/fig/{filename[:-4]}.eps', np.load(img))
```

```python
plot_img('/datax/scratch/zachy/ml-blc1/data/train/000000.npy')
```
