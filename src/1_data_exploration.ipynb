```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Function to count the number of .npy files in a directory
def file_count(path):
    count = len(glob.glob1(path, '*.npy'))
    return count
    
# Function to display (and save) dynamic spectra
def plot_img(path, save=False, clear=False):
    plt.imshow(np.load(path, allow_pickle=True), aspect='auto', interpolation='none'); plt.colorbar()
    if save:
        plt.savefig(f'{path[:-4]}.eps')
    if clear:
        plt.clf()
```

```python
# Set path to data
path = '/datax/scratch/zachy/ml-blc1/data/v1'
```

```python
# Expected count: 60,000
train_count = file_count(f'{path}/train')
print(f'Number of .npy files in train: {train_count}')

# Expected count: 40,000
test_count = file_count(f'{path}/test')
print(f'Number of .npy files in test: {test_count}')
```

```python
# Read labels.csv
train_labels = pd.read_csv(f'{path}/train/labels.csv')
test_labels = pd.read_csv(f'{path}/test/labels.csv')
```
